# Scheme joke generator
## Testing  and evaluation

The joke generation module is called scheme_joke_generator and can easily be loaded. 

```python
import scheme_joke_generator as jg
```

It generates jokes of the pattern 'I like my X like I like my Y, Z'. The algorithm is a reimplementation of this paper:

  *Petrović, S., & Matthews, D. (2013). Unsupervised joke generation from big data.
  Proceedings of the 51st Annual Meeting of the Association for Computational Linguistics (Volume 2: Short Papers), 228–232.*

Starting from a seed word, the generator calculates suitable (X,Y,Z) triples using differnt measures. Details about this can be found in the referenced paper. The triples and their scores as well as the jokes deriving from the top k triples are outputted and stored in `/output`
